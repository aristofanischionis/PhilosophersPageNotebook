```python
import pandas as pd

a = pd.read_csv("./df_ales.csv")
b = pd.read_csv("./philosophers_with_quotes.csv")
c = pd.read_csv("./philosophers_with_centuries.csv")
d = pd.read_csv("./philosophers_field.csv")
e = pd.read_csv("./simple_philosophers.csv")

final = pd.merge(a, b, on='Name')
final= pd.merge(final, c, on='Name')
final= pd.merge(final, d, on='Name')
final= pd.merge(final, e, on='Name')

final = final.loc[:,~final.columns.duplicated()]

final.head()
```

```python
final = final.drop(columns=['File name_x', 'Unnamed: 0', 'Birth/death dates_x', 'File name_y', 'Birth/death dates_y', 'Quotes'])
```

```python
final.to_csv('philosophers.csv', index=False)
```
